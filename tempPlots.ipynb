```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the dataset
file_path = 'results/RampEvapNoMacor150s_150.0s.csv'
df = pd.read_csv(file_path, encoding="latin1",sep='\t')

# Extract coordinates and temperature data
x = df['X Location (m)']
y = df['Y Location (m)']
z = df['Z Location (m)']
temp_col = df.columns[4]
temp = df[temp_col]

# Set up the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create a 3D scatter plot
# cmap='coolwarm': blue for cold, red for hot
# s=10: larger points
sc = ax.scatter(x, y, z, c=temp, cmap='coolwarm', s=10, alpha=0.6, vmin=0, vmax=800)

# Flip the Z-axis
ax.invert_zaxis()

# Labeling the axes
ax.set_xlabel('X Location (m)')
ax.set_ylabel('Y Location (m)')
ax.set_zlabel('Z Location (m)')
ax.set_title('3D Temperature Distribution (150s No Macor)')

# Add a colorbar
cbar = plt.colorbar(sc, ax=ax, shrink=0.5, aspect=5)
cbar.set_label(temp_col)

# Display the plot
plt.tight_layout()
plt.show()
```
